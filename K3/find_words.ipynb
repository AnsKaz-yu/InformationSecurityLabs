```python
import nltk
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.corpus import brown

# Загрузка корпуса Brown
nltk.download('brown')

# Получение текста из корпуса Brown
text = ' '.join(brown.words())

# Создание n-грамм из символов
n = 3  # Например, триграммы
char_ngrams = list(ngrams(text, n))

# Вычисление частот n-грамм
fdist = FreqDist(char_ngrams)


from nltk.probability import FreqDist

# Загрузка корпуса Brown
words = brown.words()

# Создание частотного распределения слов
fdist = FreqDist(words)

# Функция для предсказания слов по началу слова
def predict_words(prefix):
    return [word for word in fdist if word.startswith(prefix)]
```

```python

prefix = "th"
l = 9
predictions = predict_words(prefix)

if l:
    predictions = [word for word in predictions if len(word) == l ]
print(predictions)
```
